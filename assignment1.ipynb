3. Given two binary strings a and b, return their sum as a binary string.
For instance:
• Input: a = "11", b = "1"
• Output: "100"

```python
str1 = "11"
str2 = "1"
 

if(len(str1) > len(str2)):
    longer_str = str1
    shorter_str = str2
else:
    longer_str = str2
    shorter_str = str1


for i in range(len(longer_str) - len(shorter_str)):
    shorter_str = "0" + shorter_str


carry = 0
sum = []

for indx in range(len(longer_str) - 1, -1, -1):

    longer_d = int(longer_str[indx])
    shorter_d = int(shorter_str[indx])

    dig_sum = longer_d + shorter_d + carry

    sum.append(str(dig_sum % 2)) 
    carry = dig_sum // 2 

if carry:
    sum.append("1")


sum_str = "".join(reversed(sum))
print(sum_str)
```

4. Given a non-empty array of integers numbers, every element appears twice except for one. Find that
single one.

5. The DNA sequence is composed of a series of nucleotides abbreviated as 'A', 'C', 'G', and 'T'.
For example, "ACGAATTCCG" is a DNA sequence. When studying DNA, it is useful to identify repeated
sequences within the DNA.
Given a string s that represents a DNA sequence, return all the 10-letter-long sequences (substrings)
that occur more than once in a DNA molecule. You may return the answer in any order.

For instance:
• Input: s = "AAAAACCCCCAAAAACCCCCCAAAAAGGGTTT"
• Output: ["AAAAACCCCC","CCCCCAAAAA"]

```python
def repeated_sequence_detective(dna):

    first_time_seen = set()   
    duplicates = set()   
    
    L = 10
    if len(dna) < L:
        return []

    for i in range(len(dna) - L + 1):
        sub = dna[i:i+L]
        if sub in first_time_seen:
            duplicates.add(sub)    
        else:
            first_time_seen.add(sub)   

    return list(duplicates)

dna = "AAAAACCCCCAAAAACCCCCCAAAAAGGGTTT"
print(repeated_sequence_detective(dna)) 

dnaII="atcgatcgatggggggggggggggatcgatcgat"
print(repeated_sequence_detective(dnaII))
```
