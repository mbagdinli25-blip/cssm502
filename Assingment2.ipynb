***Data Loading and Merging***

```python
import pandas as pd
import numpy as np
```

```python
life_exp=pd.read_excel("lex.xlsx")
pop=pd.read_excel("pop.xlsx")
gdp_pcap=pd.read_excel("gdp_pcap.xlsx")
chld_mrt=pd.read_excel("child_mortality_0_5_year_olds_dying_per_1000_born.xlsx")

```

```python
display(life_exp.head(5), pop.head(5), gdp_pcap.head(5), chld_mrt.head(5))
```

```python
life_exp=life_exp.drop("geo",axis=1).set_index("name").unstack()
pop=pop.drop("geo",axis=1).set_index("name").unstack()
gdp_pcap=gdp_pcap.drop("geo",axis=1).set_index("name").unstack()
chld_mrt=chld_mrt.drop("geo",axis=1) .set_index("name").unstack()
```

```python
pop.index
```

```python
pop.index.names= ["year","country"]
life_exp.index.names= ["year","country"]
gdp_pcap.index.names= ["year","country"]
chld_mrt.index.names= ["year","country"]
```

```python
merged=pd.concat([pop,life_exp,gdp_pcap,chld_mrt], axis=1) #hocam mergele asla başaramadım o sebeple concatla yaptım
merged.columns = ["population","life expectation","gdp per capita", "child mortality"]
```

```python
asia_2000=asia[asia.index.get_level_values("year") == 2000].sort_values(["life expectation"],ascending=True)
display(asia_2000)
```

```python
upper_th = asia_2000[asia_2000["population"] > asia_2000["population"].mean()]["population"].mean()
lower_th = asia_2000[asia_2000["population"] < asia_2000["population"].mean()]["population"].mean()
```

```python
upper_th
```

```python
lower_th
```

```python
def population_category(x): #, upper_th, lower_th):
    if(x > upper_th):
        return "high"
    elif(x > lower_th):
        return "medium"
    else:
        return "low"

asia_2000["population_class"] = asia_2000["population"].apply(population_category)
    
```
